```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
df = pd.read_csv('./car data.csv')
```

```python
print(df.isnull().sum())
# No Null/nan Values
```

```python
df.describe()
```

```python
plt.style.use('seaborn')
sns.distplot(np.log1p(df['Selling_Price']),fit=st.lognorm)
```

```python
np.log1p(df['Selling_Price'])
```

```python
mean = np.mean(df['Selling_Price'])
```

```python
var = np.std(df['Selling_Price'])
```

```python
new = (df['Selling_Price']-mean)/var
new_log = np.log1p(new)
```

```python
plt.style.use('seaborn')
sns.distplot(new_log,fit=st.lognorm)
```

```python
plt.style.use('seaborn')
sns.distplot(np.log1p(df['Present_Price']),fit=st.johnsonsu)
```

```python
qualitative = [f for f in df.columns if df[f].dtypes=='object']
quantitative = [f for f in df.columns if df[f].dtypes!='object']
```
